# Explore the mandelbrot set
### Controls:

* `arrow keys` for panning
* `enter` to zoom in
* `backspace` to zoom out
* `page up` and `page down` to change resolution \*
* `a` and `z` to change colormap

```python
%matplotlib notebook
from Mandelbrot_Zoom_app import Mandelbrot_Zoom
import ipywidgets as widgets
import matplotlib.pyplot as plt

mz = Mandelbrot_Zoom()
#extra controls:
@widgets.interact
def set_cmap(cmap = widgets.Dropdown(options = plt.colormaps(), value = "magma")):
    mz.img.set_cmap(cmap)
```
